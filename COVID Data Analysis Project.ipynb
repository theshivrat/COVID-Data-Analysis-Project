```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis= 1)
```

```python
covid_df.head()
```

```python
covid_df["Date"]= pd.to_datetime(covid_df.Date, format= "%Y-%m-%d")
```

```python
##Active Cases

covid_df["Active_Cases"]= covid_df["Confirmed"] - covid_df["Cured"]-covid_df["Deaths"]
covid_df.tail()
```

```python
 statewise = pd.pivot_table(covid_df, values= ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise = statewise.sort_values("Confirmed", ascending= False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
statewise.drop(["Maharashtra***", "Madhya Pradesh***", "Bihar****","Karanataka"], inplace= True)
```

```python
#TOP 10 ACTIVE CASES STATE
top_10_ActiveCases= covid_df.groupby("State/UnionTerritory").max()[["Active_Cases", "Date"]].sort_values("Active_Cases", ascending = False).reset_index().head(10)
```

```python
plt.bar(top_10_ActiveCases["State/UnionTerritory"], top_10_ActiveCases["Active_Cases"], linewidth = 10 )
plt.xlabel("State")
plt.ylabel("Active Cases")
plt.title(" Top 10 State by Active Cases")
plt.xticks(fontsize = 8, rotation = 90)
plt.show()
```

```python
#TOP 10 DEATHS REPORTED STATES

top_10_Deaths= statewise.sort_values("Deaths",ascending = False).head(10)
```

```python
plt.bar(top_10_Deaths.index, top_10_Deaths["Deaths"], linewidth = 10 )
plt.xlabel("State")
plt.ylabel("No. of Deaths")
plt.title(" Top 10 Most Deaths Reported States")
plt.xticks(fontsize = 8, rotation = 90)
plt.show()
```

```python
Vaccination = vaccine_df.drop(["Sputnik V (Doses Administered)", "AEFI", "18-44 Years (Doses Administered)", "45-60 Years (Doses Administered)","60+ Years (Doses Administered)" ], axis = 1)
Vaccination.head()
```

```python
# Male VS Female Vaccination

MaleV = Vaccination["Male(Individuals Vaccinated)"].sum()
FemaleV = Vaccination["Female(Individuals Vaccinated)"].sum()
plt.pie([MaleV, FemaleV], labels = ["Male", "Female"], autopct = '%1.1f%%')
plt.title("Male and Female Vaccination")
plt.show()
```

```python
# remove rows where state = India
stateindia = Vaccination[Vaccination["State"] == "India"].index
Vaccination.drop(stateindia, axis= 0, inplace = True)
Vaccination.head()
```

```python
#Most Vaccinated State

max_vacstate = Vaccination.groupby("State")["Total Individuals Vaccinated"].sum().to_frame("Total")
max_vacstate = max_vacstate.sort_values("Total", ascending = False)[:5]

plt.bar(max_vacstate.index, max_vacstate["Total"])
plt.xlabel("State")
plt.ylabel("Vaccination")
plt.xticks(rotation = 90, fontsize = 10)
plt.title("Most Vaccinated State")
```

```python
#Least Vaccinated States

leastVac = Vaccination.groupby("State")["Total Individuals Vaccinated"].sum().to_frame("Total")
leastVac = leastVac.sort_values("Total", ascending = False).tail(5)

plt.bar(leastVac.index, leastVac["Total"])
plt.xlabel("State")
plt.ylabel("Vaccination")
plt.xticks(rotation = 90, fontsize = 10)
plt.title("Most Vaccinated State")
```
